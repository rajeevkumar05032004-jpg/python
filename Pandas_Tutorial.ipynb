#### A Series in pandas is a one-dimensional labeled array capable of holding data of any type (integer, float, string, etc.). It is similar to a column in a DataFrame or a one-dimensional NumPy array but comes with labels (called the index) for each element.

### From List

```python
data = [10, 20, 30, 40]
series = pd.Series(data)
print(series)
```

### From a List with Custom Index

```python
series = pd.Series(data, index=['a', 'b', 'c', 'd'])
print(series)
```

### From a Dictionary

```python
data = {'a': 100, 'b': 200, 'c': 300}
series = pd.Series(data)
print(series)
```

### From a NumPy Array

```python
import numpy as np
array = np.array([1.1, 2.2, 3.3, 4.4])
series = pd.Series(array, index=['P', 'Q', 'R', 'S'])
print(series)
```

### Accessing Data in a Series

### 1. Access by Index

```python
print(series['S'])  # Access by label
```

### 2. Access by Position

```python
print(series.iloc[2])  # Access by position
```

### 3. Slicing

```python
print(series[1:3])  # Slice by position
```

### 1. Vectorized Operations

```python
print(series + 10)  # Add 10 to each element
```

```python
print(series[series > 2])  # Filter elements greater than 2
```

## Dataframe
### A DataFrame is a two-dimensional, tabular data structure in pandas, a popular Python library for data analysis. It is similar to an Excel spreadsheet or a SQL table, where data is stored in rows  and columns.

## Key Features of a DataFrame:
### 1. Labeled Rows and Columns – Each column has a name, and each row has an index.

### 3. Supports Operations – Filtering, aggregation, merging, and other data manipulation tasks.

### 4. Missing Data Handling – Built-in methods to deal with missing values.

### dropna() : used to remove rows or columns with missing values

```python
import numpy as np
data = {'A': [1, 2, np.nan], 'B': [4, np.nan, 6], 'C': [7, 8, 9]}
df = pd.DataFrame(data)
print(df)
print("----------------")
result = df.dropna()
print(result)
print("----------------")
result = df.dropna(axis=0)  # Drop rows with null value 
print(result)
print("----------------")
result = df.dropna(axis=1)  # Drop column with null value  
print(result)
```

### astype(): It is used to explicitly cast a Pandas object (like a DataFrame or Series) to a specified data type (dtype).

```python
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': ['4', '5.63', '6.7']
})
print(df)
# Convert column 'A' to float
df['A'] = df['A'].astype(float)
print(df)
```

### Drop Duplicates : Remove duplicate rows

```python
# Sample DataFrame
data = {'A': [1, 2, 2, 3, 4, 4],
        'B': ['a', 'b', 'b', 'c', 'd', 'd']}
df = pd.DataFrame(data)
print(df)
print("---------------")
# Drop duplicates based on all columns
df_no_duplicates = df.drop_duplicates()
print(df_no_duplicates)
```

### Filtering a DataFrame Based on a Condition:

```python
# Creating a sample DataFrame
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 22, 28],
    'Salary': [50000, 70000, 45000, 65000]
}

df = pd.DataFrame(data)

# Filtering rows where Age is greater than 25
df_filtered2 = df[df['Age'] > 25]

print(df_filtered2)
```

### Aggregating and Grouping Data in Pandas
### ✅ 1. Grouping Data with groupby()
### The groupby() function in pandas is used to group data based on one or more columns, and then we can apply aggregation functions like sum, mean, count, etc.

```python
# Sample DataFrame
data = {
    'Department': ['HR', 'IT', 'HR', 'IT', 'Finance', 'Finance', 'HR'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    'Salary': [50000, 70000, 45000, 80000, 60000, 55000, 52000]
}

df = pd.DataFrame(data)
print(df)
```

### Grouping by 'Department' and calculating total salary

```python
df_grouped = df.groupby('Department')['Salary'].sum()
print(df_grouped)
```

### Grouping by Department and applying multiple aggregations

```python
df_grouped = df.groupby('Department')['Salary'].agg(['sum', 'mean', 'count'])
print(df_grouped)
```

### Grouping by Multiple Columns

```python
df_grouped = df.groupby(['Department', 'Employee']).sum()
print(df_grouped)
```

### Using transform() to Add Grouped Data to Original DataFrame

```python
df['Avg_Salary_by_Dept'] = df.groupby('Department')['Salary'].transform('mean')
print(df)
```

## Pivot Table
### df.pivot_table() in Pandas
### The pivot_table() function in pandas is used to summarize and aggregate data by reshaping it
### into a table format. It is particularly useful for multi-dimensional analysis, similar to Excel pivot tables.
### syntax: df.pivot_table(values, index, columns, aggfunc)
### values ->	The column to aggregate (e.g., 'Salary')
### index ->	The row labels (grouping by this column)
### columns -> The column labels (optional, used to pivot data)
### aggfunc -> The aggregation function (default is mean())

```python
data = {
    'Department': ['HR', 'IT', 'HR', 'IT', 'Finance', 'Finance', 'HR'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    'Salary': [50000, 70000, 45000, 80000, 60000, 55000, 52000]
}

df = pd.DataFrame(data)
print(df)
print("---------------------------")
# Creating a pivot table: Average Salary per Department
pivot = df.pivot_table(values='Employee', index='Department', aggfunc='count')

print(pivot)
```

### Pivot Table with Multiple Aggregations

```python
pivot = df.pivot_table(values='Salary', index='Department', aggfunc=['mean','count','sum'])
print(pivot)
```

### Pivot Table with Multiple Indexes

```python
pivot = df.pivot_table(values='Salary', index=['Department', 'Employee'], aggfunc='sum')
print(pivot)
```

### Pivot Table with columns Parameter

```python
pivot = df.pivot_table(values='Salary', index='Department', columns='Employee', aggfunc='sum', fill_value=0)
print(pivot)
```

### 🔹 Merging & Joining Data in Pandas
#### Concatenating DataFrames (pd.concat())
#### concat() is used to stack DataFrames vertically (rows) or horizontally (columns).

```python

df1 = pd.DataFrame({
    'Emp_ID': [101, 102, 103, 104],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Department': ['HR', 'IT', 'HR', 'Finance']
})

df2 = pd.DataFrame({
    'Emp_ID': [101, 102, 105, 106],
    'Salary': [50000, 70000, 65000, 60000]
})
print(df1)
print("---------------------")
print(df2)
print("---------------------")
```

```python
df_concat = pd.concat([df1, df2], axis=0)  # axis=0 → Row-wise
print(df_concat)
```

```python
df_concat = pd.concat([df1, df2], axis=1)  # axis=1 → Column-wise
print(df_concat)
```

### Merging on Emp_ID
#### INNER JOIN (how='inner')

```python
df_merged = pd.merge(df1, df2, on='Emp_ID', how='inner')
print(df_merged)
```

#### LEFT JOIN (how='left')

```python
df_left = pd.merge(df1, df2, on='Emp_ID', how='left')
print(df_left)
```

#### RIGHT JOIN (how='right')

```python
df_right = pd.merge(df1, df2, on='Emp_ID', how='right')
print(df_right)
```

#### OUTER JOIN (how='outer')

```python
df_outer = pd.merge(df1, df2, on='Emp_ID', how='outer')
print(df_outer)
```

### 🔹 Data Visualization in Pandas
##### 1.Line Plot	-> df.plot(kind='line')	Trends over time
##### 2.Bar Chart	-> df.plot(kind='bar')	Comparing categories
##### 3.Histogram	-> df.plot(kind='hist')	Distribution of data
##### 4.Scatter Plot -> df.plot(kind='scatter')	Relationship between variables
##### 5.Box Plot	-> sns.boxplot()	: Outlier detection
##### 6.Pie Chart	-> df.plot(kind='pie')	Proportions of categories

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Enable inline plotting (for Jupyter Notebooks)
%matplotlib inline  
```

```python
data = {
    'Emp_ID': [101, 102, 103, 104, 105],
    'Age': [25, 30, 35, 35,45],
    'Salary': [50000, 60000, 75000, 80000, 62000],
    'Department': ['HR', 'IT', 'HR', 'Finance', 'IT']
}

df = pd.DataFrame(data)
print(df)
```

### Line Plot : plot.line()

```python
df.plot(x='Age', y='Salary', kind='line',marker='o',linestyle='--',color='r')
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid()
plt.show()
```

### Bar Plot : plot.bar()

```python
df.plot(x='Department', y='Salary', kind='bar', color='blue', edgecolor='black')
plt.title("Salary by Department")
plt.xlabel("Department")
plt.ylabel("Salary")
plt.show()
```

### Histogram : plot.hist()

```python
df['Age'].plot(kind='hist', bins=4, color='purple', edgecolor='black')
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.show()
```

### Scatter Plot : plot.scatter()

```python
df.plot(kind='scatter', x='Age', y='Salary', color='red')
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid()
plt.show()
```

### Box Plot : sns.boxplot()

```python
sns.boxplot(x=df['Salary'], color='orange')
plt.title("Salary Distribution")
plt.show()
```

### Pie Chart : plot.pie()

```python
df['Department'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['red', 'blue', 'green'])
plt.title("Department Distribution")
plt.ylabel("")  # Hide y-axis label
plt.show()
```
