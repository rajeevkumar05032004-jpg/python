### What are Python modules and packages?
Modular programming refers to the process of breaking a large, unwieldy programming task into separate, smaller, more manageable subtasks or modules. Individual modules can then be cobbled together like building blocks to create a larger application.

*  <u>Simplicity</u>: Rather than focusing on the entire problem at hand, a module typically focuses on one relatively small portion of the problem. If you’re working on a single module, you’ll have a smaller problem domain to wrap your head around. This makes development easier and less error-prone.
   
* <u>Maintainability</u>: Modules are typically designed so that they enforce logical boundaries between different problem domains. If modules are written in a way that minimizes interdependency, there is decreased likelihood that modifications to a single module will have an impact on other parts of the program. (You may even be able to make changes to a module without having any knowledge of the application outside that module.) This makes it more viable for a team of many programmers to work collaboratively on a large applicatioode.
* <u>Reusability</u>: Functionality defined in a single module can be easily reused (through an appropriately defined interface) by other parts of the application. This eliminates the need to duplicate code.e.

### Difference between copy and view:

### 🔁 Copy
A copy creates a new array object with its own separate data.

Changes made to the copy do not affect the original array.

### 👀 View
A view is a shallow copy: it’s a new array object that shares the same data as the original.

Changes made to the view affect the original array (and vice versa).

```python
original = np.array([1,2,3,4,5,6])
print(original)
copy=original.copy()
print(copy)
copy[0]=100
print(copy)
print(original)
```

```python
original1 = np.array([1,2,3,4,5,6])
print(original1)
view=original1.view()
print(view)
view[0]=32
print(view)
print(original1)
```

```python
arr8=np.array([1,2,3,4,5,6])
e=arr8.copy()
f=arr8.view()
print(e.base)
print(f.base)
```

### Median
### numpy.median()

Compute the median of the given data (array elements) along the specified axis.

numpy.median(arr, axis = None)

Median Formula ={(n+1)/2}th when n is odd

Median Formula ={(n)/2)th}

### Median Formula Example:-
#### List of numbers: 4, 10, 7, 15, 2. Calculate the median.

Solution: Let us arrange the numbers in ascending order.

In ascending order, the numbers are: 2,4,7,10,15

Therefore, calculation of the median will be as follows,

There are a total of 5 numbers. Median is (n+1)/2 th value.

Thus, the Median is (5+1)/2 th value.

Median = 3rd value.

The 3rd value in list 2, 4, 7, 10, 15 is 7.

Median will be –

Thus, the Median is 7.

```python
a = np.array([4,10,7,15,2])
print(np.median(a))
```

# Variance :
## numpy.var()
<img src="Variance.jpg">
Compute the variance of the given data (array elements) along the specified axis.

Variance is the average of squared deviations, i.e., mean(abs(x - x.mean())**2).

In other words, the standard deviation is the square root of variance.

numpy.var(arr, axis = None)

```python
# 1D array  
arr =np.array([20, 2, 7, 1, 34])  
  
print("arr : ", arr)  
print("variance of arr : ", np.var(arr))  
```
